### LaVague Launch

We will now download the default configuration files for running LaVague with deepseek-coder 

```python
!wget https://raw.githubusercontent.com/lavague-ai/LaVague/main/examples/local/deepseek-coder.py
!wget https://raw.githubusercontent.com/lavague-ai/LaVague/main/tests/hf.txt
```

```python
!lavague-launch --file_path hf.txt --config_path deepseek-coder.py
```

You can now click on the public (if you are using Google Colab) or local URL to open the Gradio in your browser.

### LaVague Build

We can alternatively use the `lavague-build` command to generate a python script with the Selenium code needed to perform your desired action.

```python
!lavague-build --file_path hf.txt --config_path deepseek-coder.py
```

This creates a script with the Python code generated by the LLM to perform your desired action in your current directory named `hf_deepseek-coder.py`.

You can now inspect the code or execute it locally!
